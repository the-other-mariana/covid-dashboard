```python
sub_covid_df = covid_df[["ID_REGISTRO", "SEXO", "ENTIDAD_RES", "MUNICIPIO_RES", "EDAD", "FECHA_DEF"]]
sub_covid_df.head()
```

```python
sub_covid_df.groupby(["ENTIDAD_RES", "SEXO"], as_index=False)["ID_REGISTRO"].count() # keep data as df format instead of series
```

```python
sub_count = sub_covid_df.groupby(["ENTIDAD_RES", "SEXO"], as_index=False)["ID_REGISTRO"].count()
sub_count.rename(columns={"ID_REGISTRO": "COUNT"}, inplace=True)
sub_count
```

```python
# from subcount, get only the rows where the column COUNT is > 5
sub_count[sub_count["COUNT"] > 5]
```

```python
sub_count[sub_count["COUNT"] > 5].groupby("ENTIDAD_RES")["COUNT"].sum() # by state, num of covid infected
# apply [COUNT].sum() to my groups (by state)
```

```python
# from subcount, give me the ones with > 10 counts and where their entidad is DF
sub_count[(sub_count["COUNT"] > 10) & (sub_count["ENTIDAD_RES"] == "DF")]
```

```python
# split in groups (by entidad) and for each group, grab count and sum all its counts (2 entries = 172)
sub_count[(sub_count["COUNT"] > 10) & (sub_count["ENTIDAD_RES"] == "DF")].groupby("ENTIDAD_RES")["COUNT"].sum()
```

```python
sub_count[(sub_count["COUNT"] > 10) | (sub_count["ENTIDAD_RES"] == "DF")]
```

```python
# fom subcount, show me the ones with > 10 or where their entidad is equal to either df, nl, or jc
sub_count[(sub_count["COUNT"] > 10) | (sub_count["ENTIDAD_RES"].isin(["DF", "NL", "JC"]))].head(20)
```
